```python
import numpy as np
import os

files = os.listdir("data/")
```

```python
try:
    del X_train, y_train
    del X_test, y_test
    print('Cleared previous data')
except:
    pass

test_set = os.listdir("data/test_set/")
train_set = os.listdir("data/training_set/")

all_image_paths = []

# Gets path of every data point
for root, dirs, files, in os.walk("data"):
    for file in files:
        if file.endswith(".jpg"):
            all_image_paths.append(os.path.join(root, file))

print(f"Amount of data points:", len(all_image_paths))

test_image_paths = []

for root, dirs, files, in os.walk("data/test_set"):
    for file in files:
        if file.endswith(".jpg"):
            test_image_paths.append(os.path.join(root, file))

print(f"Amount of test data points:", len(test_image_paths))

training_image_paths = []

for root, dirs, files, in os.walk("data/training_set/"):
    for file in files:
        if file.endswith('.jpg'):
            training_image_paths.append(os.path.join(root, file))

print(f"Amount of training data points:", len(training_image_paths))

print(training_image_paths[1])
```

```python
import cv2

# print(training_image_paths[1])
# cv2.resize(img, )
image = cv2.imread('data/training_set/cats/cat.9.jpg')

new_width = 256
new_height = 256

resized_image = cv2.resize(image, (new_width, new_height))

cv2.imshow("Image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#####################################
# Create folders for resized_images, loop through the testing data
# and write data into folders between cats and dogs
#####################################
os.makedirs("resized_data/test_set/cats/", exist_ok=True)
os.makedirs("resized_data/test_set/dogs/", exist_ok=True)

# Define the cat and dog counter
dog_num, cat_num = 1, 1

for i in range(len(test_image_paths)):
    image = cv2.imread(test_image_paths[i])

    if "cat" in test_image_paths[i]:
        resized_image = cv2.resize(image, (new_width, new_height))
        cv2.imwrite(f"resized_data/test_set/cats/cat.{cat_num}.jpg", resized_image) 
        cat_num += 1
    elif "dog" in test_image_paths[i]:
        resized_image = cv2.resize(image, (new_width, new_height))
        cv2.imwrite(f"resized_data/test_set/dogs/dog.{dog_num}.jpg", resized_image)      
        dog_num += 1
```

```python
#####################################
# Create folders for resized_images(if not made already), loop through the training data
# and write data into folders between cats and dogs
#####################################
os.makedirs("resized_data/training_set/cats/", exist_ok=True)
os.makedirs("resized_data/training_set/dogs/", exist_ok=True)

# Define the cat and dog counter
dog_num, cat_num = 1, 1

for i in range(len(training_image_paths)):
    image = cv2.imread(training_image_paths[i])

    if "cat" in training_image_paths[i]:
        resized_image = cv2.resize(image, (new_width, new_height))
        cv2.imwrite(f"resized_data/training_set/cats/cat.{cat_num}.jpg", resized_image)
        cat_num += 1
    elif "dog" in training_image_paths[i]:
        resized_image = cv2.resize(image, (new_width, new_height))
        cv2.imwrite(f"resized_data/training_set/dogs/dog.{dog_num}.jpg", resized_image)
        dog_num += 1
```
