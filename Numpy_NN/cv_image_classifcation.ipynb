Given we don't need to do any conversions, lets define useable and easily callable variables X_train, y_train, X_test, y_test.

```python
####################################
# 1. Define cat(0) and dog(1) labels
# 2. Loop through each class
# 3. Loop through each file in the class 
# 4. Read in the files using cv2
# 5. Append the files to test_data
# 6. Append the label based on class name
# 7. Turn test_data into a numpy array
# 8. Turn y_test_data into numpy array
####################################

labels = {"cats": 0, "dogs": 1}

test_data = []
y_test_data = []

rsz_test_dir = "resized_data/test_set"

for class_name in os.listdir(rsz_test_dir):
    class_path = os.path.join(rsz_test_dir, class_name)

    if not os.path.isdir(class_path):
        continue

    for file in os.listdir(class_path):
        img_path = os.path.join(class_path, file)

        img = cv2.imread(img_path)
        if img is None:
            continue
        test_data.append(img)
        y_test_data.append(labels[class_name])

X_test = np.array(test_data)
y_test = np.array(y_test_data)
```

```python
from random import randrange

# Pulled from Stanford Online CSC231 Course
# grad_check_sparse compares the computed gradient to a numerically approximated gradient

def grad_check_sparse(f, x, analytic_grad, num_checks=10, h=1e-5):
    """
    sample a few random elements and only return numerical
    in this dimensions.
    """

    for i in range(num_checks):
        ix = tuple([randrange(m) for m in x.shape])

        oldval = x[ix]
        x[ix] = oldval + h  # increment by h
        fxph = f(x)  # evaluate f(x + h)
        x[ix] = oldval - h  # increment by h
        fxmh = f(x)  # evaluate f(x - h)
        x[ix] = oldval  # reset

        grad_numerical = (fxph - fxmh) / (2 * h)
        grad_analytic = analytic_grad[ix]
        rel_error = abs(grad_numerical - grad_analytic) / (
            abs(grad_numerical) + abs(grad_analytic)
        )
        print(
            "numerical: %f analytic: %f, relative error: %e"
            % (grad_numerical, grad_analytic, rel_error)
        )
```
