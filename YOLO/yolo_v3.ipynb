```python
import torch
import torch.nn as nn
import torchvision
import torchvision.datasets as datasets
import torchvision.transforms as transforms

import numpy as np
```

```python
class DarkNetBlock(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        mid_channels = in_channels // 2
        self.seq1 = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=1, stride=1, padding=0, bias=False),
            nn.BatchNorm2d(mid_channels),
            nn.LeakyReLU(0.1, inplace=True),
            nn.Conv2d(mid_channels, in_channels, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(in_channels),
            nn.LeakyReLU(0.1, inplace=True)
        )

    def forward(self, x):
        return x + self.seq1(x)
```

```python
class Darknet53(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(64, 32, kernel_size=3, stride=1, padding=0)
        self.layer1 = self._make_layers(DarkNetBlock)
        self.conv3 = nn.Conv2d(128, 64, kernel_size=3, stride=1, padding=0)
        self.block2 = Darknet53()



    def _make_layers(self, block, out_channels, num_blocks):
        layers = []

        for block in num_blocks:
            layers.append(block(self.in_channels))
            self.in_channels = out_channels

        return nn.Sequential(*layers)
```
