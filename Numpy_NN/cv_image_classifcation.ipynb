Given we don't need to do any conversions, lets define useable and easily callable variables X_train, y_train, X_test, y_test.

```python
####################################
# 1. Define cat(0) and dog(1) labels
# 2. Loop through each class
# 3. Loop through each file in the class 
# 4. Read in the files using cv2
# 5. Append the files to test_data
# 6. Append the label based on class name
# 7. Turn test_data into a numpy array
# 8. Turn y_test_data into numpy array
####################################

labels = {"cats": 0, "dogs": 1}

test_data = []
y_test_data = []

rsz_test_dir = "resized_data/test_set"

for class_name in os.listdir(rsz_test_dir):
    class_path = os.path.join(rsz_test_dir, class_name)

    if not os.path.isdir(class_path):
        continue

    for file in os.listdir(class_path):
        img_path = os.path.join(class_path, file)

        img = cv2.imread(img_path)
        if img is None:
            continue
        test_data.append(img)
        y_test_data.append(labels[class_name])

X_test = np.array(test_data)
y_test = np.array(y_test_data)
```

```python
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

os.makedirs("numpy_data", exist_ok=True)

np.save("numpy_data/X_test.npy", X_test)
np.save("numpy_data/y_test.npy", y_test)
```

Made new directory consisting of our numpy data. Saved the numpy arrays into the new directory.

```python
train_data = []
y_train_data = []


rsz_train_dir = "resized_data/training_set"

for class_name in os.listdir(rsz_train_dir):
    class_path = os.path.join(rsz_train_dir, class_name)

    if not os.path.isdir(class_path):
        continue

    for file in os.listdir(class_path):
        img_path = os.path.join(class_path, file)

        img = cv2.imread(img_path)
        if img is None:
            continue

        train_data.append(img)
        y_train_data.append(labels[class_name])

X_train = np.array(train_data)
y_train = np.array(y_train_data)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")

np.save("numpy_data/X_train.npy", X_train)
np.save("numpy_data/y_train.npy", y_train)
```
